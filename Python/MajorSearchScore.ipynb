```python
# 수강 과목 데이터 불러오기
filename = '/content/drive/My Drive/Major/ClassNew.txt'
Class = pd.read_csv(filename, sep = "\t")
Class = Class.replace('학문의융합(전공탐색교양)', '전공탐색')
Class.head(50)
```

```python
MergeTable = pd.merge(Result, Class, how='inner', on=None)
MergeTable.head(50)
```

```python
# 해당 학년도 제외 삭제
MergeTable = MergeTable[MergeTable['학년도']==MergeTable['기준연도']]

# 전공탐색 수업 제외 삭제
MergeTable = MergeTable[MergeTable['이수구분']=='전공탐색']
MergeTable.head(100)
# MergeTable.to_excel('MergeTable.xlsx')
```

```python
# 1지망 전공선택 점수

FirstRank = MergeTable.copy()
FirstRank['전공선택점수1']=0
FirstRank.loc[(FirstRank['개설_학과'] == FirstRank['1지망']), '전공선택점수1']=400
FirstRank.head(50)
```

```python
# 1지망 전공선택 점수

FirstRank = MergeTable.copy()
FirstRank['전공선택점수1']=0

FirstRank.loc[(FirstRank['개설_학과'] == FirstRank['1지망']), '전공선택점수1']=400

# ID 기준 그룹화
FirstRankGroup = FirstRank.groupby('ID').전공선택점수1.sum().reset_index()
FirstRankGroup.head(50)
```

```python
# 1지망 전공선택 점수

FirstRank = MergeTable.copy()
FirstRank['전공선택점수1']=0

FirstRank.loc[(FirstRank['개설_학과'] == FirstRank['1지망']), '전공선택점수1']=400

# ID 기준 그룹화
FirstRankGroup = FirstRank.groupby('ID').전공선택점수1.sum().reset_index()
FirstRankGroup.loc[(FirstRankGroup.전공선택점수1 >= 400), '전공선택점수1']=400
FirstRankGroup.to_excel('FRG.xls')

# 1지망 점수와 머지테이블 머지
Result1 = pd.merge(Result, FirstRankGroup, how='inner', on=None)
Result1
```
