```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
filename = '/content/drive/MyDrive/Major/1123/majorscore.csv'
major = pd.read_csv(filename, encoding='utf-8')
major = major.drop(major.columns[0],axis=1)
major.head(50)
```

```python
filename = '/content/drive/MyDrive/Major/1123/extracurricular.csv'
extracurricular = pd.read_csv(filename, encoding='utf-8')
extracurricular = extracurricular.drop(extracurricular.columns[0],axis=1)
extracurricular.head(50)
```

```python
Result = pd.merge(major, extracurricular, on = 'ID', how = 'left')
Result.head(50)
```

```python
filename = '/content/drive/MyDrive/Major/1123/grade.csv'
grade = pd.read_csv(filename, encoding='utf-8')
grade = grade.drop(labels='대학',axis=1)
grade = grade.drop(labels='학과(전공)',axis=1)
grade = grade.drop(labels='입학학년도',axis=1)
grade.head(50)
```

```python
Result = pd.merge(Result, grade, on = 'ID', how = 'left')
Result.head(50)
```
